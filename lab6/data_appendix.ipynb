```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('1_KUJAWSKO-POMORSKIE.csv', index_col=0)

df
```

Opis metadanych
Dane to wyniki oceny odkurzaczy w jednej z sieci sklepów z danego województwa oraz dane kupujących. Są opisane przez 5 kolumn: 'Dni od zakupu', 'Marka', 'Wiek kupującego', 'Płeć kupującego', 'Ocena'. 'Dni od zakupu' to int z zakresu 1-17, 'Marka' to nazwa jednej z pięciu marek odkurzaczy, 'Wiek kupującego' to int oznaczający wiek kupującego (w niektórych rekordach wartość NaN), 'Płeć kupującego' to symbol oznaczający płeć kupującego przyjmujący wartości ze zbioru {K, M, bd.}, 'Ocena' to float z zakresu 0-5 co 0.5

Korzystając z biblioteki seaborn utworzyłem 5 histogramów dla każdej z kolumn, gdzie wysokość słupka była proporcjonalna do ilości rekordów z daną wartością w zbiorze danych.

```python
for col in df.columns:
    plt.figure()
    sns.histplot(data=df, x=col)
```

Następnie utworzyłem wykresy słupkowe przedstawiające zależności pomiędzy poszczególnymi kolumnami. W przypadku, gdy obie kolumny nie przedstawiały wartości numerycznych, wyrysowałem countploty, które wyliczały ilość wystąpień danego rekordu. Wykresy słupkowe oprócz przedstawienia wartości średniej zaznaczają również przedział ufności.

```python
for col in df.columns:
    if col != 'Wiek kupującego':
        for col2 in df.columns:
            if col == col2:
                continue
            else:
                if df[col].dtype == object and df[col2].dtype == object:
                    plt.figure()
                    sns.countplot(data=df, hue=col, y=col2)
                else:
                    plt.figure()
                    sns.barplot(data=df, x=col, y=col2)
```
